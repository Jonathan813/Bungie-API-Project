```python
### Destiny Info ###

import requests
import pandas as pd

# API key
HEADERS = {"X-API-Key":'xxxxxxxxxxxxxxxxxxxxxxxxxxxxx'}

# Clan
clanId = 2110109

# Get Clan roster
url = "https://www.bungie.net/Platform/GroupV2/{}/Members/".format(clanId)
r = requests.get(url, headers=HEADERS);
player = r.json()['Response']['results']

# FOR TESTING INDIVIDUAL MEMBERS
#playerTemp = r.json()['Response']['results'][11]
#playerTemp

# Store display names
displayNames = []

# Retrieve display names from clan roster
for i in range(len(player)):
    playerTemp = r.json()['Response']['results'][i]
    if 'bungieNetUserInfo' in playerTemp:
        player = playerTemp['bungieNetUserInfo']['displayName']
        print(player)
    displayNames.append(player)

displayNames
```

```python
# Store blizzard names
blizzardNames = []

# Get clan members' Blizzard names
def GetBlizzardName(displayNames):
    for eachName in displayNames:
        url = "https://www.bungie.net/Platform/User/SearchUsers/?q={}".format(eachName)
        r = requests.get(url, headers=HEADERS);
        try:
            data = r.json()['Response'][0]['blizzardDisplayName']
            data = data.replace('#', '%23')
            blizzardNames.append(data)
        except:
            # Continue to next iteration.
            continue

GetBlizzardName(displayNames)
print(blizzardNames)
```

```python
# Get membershipIDs
membershipIDs = []

# Acquire Player Info
for name in blizzardNames:
    url = "https://www.bungie.net/Platform/Destiny2/SearchDestinyPlayer/4/{}/".format(name)
    r = requests.get(url, headers=HEADERS);
    membershipID = r.json()['Response'][0]['membershipId']
    membershipIDs.append(membershipID)        

print(membershipIDs)
```

```python
data_json_storage = []
all_players_data = []

# Get Player Data
for ID in membershipIDs:
    url = "https://www.bungie.net/Platform/Destiny2/4/Account/{}/Stats/".format(ID)
    r = requests.get(url, headers=HEADERS);
    data_json_storage.append(r.json())
    count = membershipIDs.index(ID)
    print(count)
    data = data_json_storage[count]['Response']['mergedAllCharacters']['results']['allPvE']['allTime']
    
    PVE_data = []
    for i in data:
        x = data[i]['basic']['value']
        PVE_data.append(x)
    
    # PVE col headers
    PVE_headers = data_json_storage[0]['Response']['mergedAllCharacters']['results']['allPvE']['allTime'].keys()
    
    # Create dictionary of player data
    player_dict = dict(zip(PVE_headers, PVE_data))
    player_dict
    all_players_data.append(player_dict)

    print("ID: {}".format(ID))
    print("Member Index: {}".format(membershipIDs.index(ID)))
    
    print("Data Index: {}".format(all_players_data.index(player_dict)))
    
player_dict = dict(zip(blizzardNames, all_players_data))
player_dict
```

```python
nameList = []
for i in player_dict:
    nameList.append(i)

len(nameList)
nameList
```

```python
# Make Dataframe
df = pd.DataFrame(all_players_data)
# Remove duplicate rows
df = df.drop(df.index[8])
df = df.drop(df.index[41])
```
